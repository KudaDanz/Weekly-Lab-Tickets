## Part 4: Building Your Finance Tracker Foundation (15 minutes)

### Start Your Semester Project

Create a new notebook called "Personal_Finance_Tracker" and build the foundation:

````python
print("=== Personal Finance Tracker v0.1 ===")
print("Week 3: Smart Budget Decisions")
print()

# User setup
name = input("Enter your name: ")
monthly_income = float(input("Enter your monthly income: $"))

print(f"\nWelcome to your finance tracker, {name}!")
print(f"Monthly income: ${monthly_income:.2f}")

# This week's feature: Expense categorisation
print("\n--- Expense Entry ---")
expense_amount = float(input("Enter an expense: $"))
expense_description = input("What was this expense for? ")


# Apply your decision-making skills here
description = expense_description.lower()
# .lower()means to put the output in lowercase e.g. users'-input = "Food" it will be taken as "food"

# Expense categorisation
# in checks if something is inside a group. e.g. is "food" in description
if description in ["rent", "food", "groceries", "transport"]:
    category = "Need"
elif description in ["shopping", "movie", "games", "clothes"]:
    category = "Want"
else:
    category = "Other"

# Budget percentage calculation
percentage = (expense_amount / monthly_income) * 100

# Display results
print("\n--- Expense Summary ---")
print(f"Expense: ${expense_amount:.2f}")
print(f"Description: {expense_description}")
print(f"Category: {category}")
print(f"This expense is {percentage:.2f}% of your monthly income.")

# Helpful feedback
if percentage > 50:
    print("Warning: This expense takes up a large part of your income.")
elif percentage > 20:
    print("Be careful, this is a significant expense.")
else:
    print("Good job! This expense is within a reasonable range.")

# Categorise the expense and give feedback
# This becomes the foundation you'll build on each week!
````
