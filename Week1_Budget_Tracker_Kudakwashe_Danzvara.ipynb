## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- üè∑Ô∏è Expense categories (food, transport, entertainment)
- üìä Show percentage of income spent
- ‚úÖ Input validation (no negative numbers)
- üìù Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add:Show percentage of income spent"
enhancement_prompt = "Can you please assist me to write a simplified python program that; \n\n1)allows users to input their monthly income, \n2)allows users to input up to 3 expenses, \n3)calculates remaining budget and \n4)shows a warning if budget in negative \n5)shows percentage of income spent"

print(f"üöÄ ENHANCEMENT: {enhancement_choice}")
print()
print(f"üìä ENHANCEMENT PROMPT: {enhancement_prompt}")
```

```python
# ENHANCED BUDGET TRACKER
# [# Ask user for monthly income
# Monthly Budget Calculator

# Ask user for monthly income
income = float(input("Enter your monthly income: "))

total_expenses = 0

# Allow user to enter up to 3 expenses
for i in range(1, 4):
    expense = input(f"Enter expense {i} (or press Enter to skip): ")

    if expense == "":
        break  # Stop if user presses Enter

    total_expenses += float(expense)

# Calculate remaining budget
remaining_budget = income - total_expenses

# Calculate percentage spent
if income > 0:
    percentage_spent = (total_expenses / income) * 100
else:
    percentage_spent = 0

# Display results
print("\n----- Budget Summary -----")
print(f"Total Expenses: {total_expenses}")
print(f"Remaining Budget: {remaining_budget}")
print(f"Percentage of Income Spent: {percentage_spent:.2f}%")

# Warning if budget is negative
if remaining_budget < 0:
    print("WARNING: Your budget is negative!")
```
